## I. Estimation

### I.1 Non-parametric estimation

#### I.1.1 Survival Function

**Kaplan-Meier Estimator**

Non-parametric estimation of the survival function using maximum likelihood.


\begin{equation}
\hat{S}(t) = \sum_{i; t_i <t} \dfrac{n_i - d_i}{n_i}
\end{equation}

>with $n_i$ the number of individuals who has not experienced the event at time $t_i$

>and $d_i$ the number of individuals that experience the event at time $t_i$

```python
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
kmf.fit(t_train, e_train)
```

```python
fig, ax = plt.subplots()
kmf.plot(ax=ax)
ax.set_xlabel('t in days')
ax.set_ylabel('S_hat(t)')
ax.set_title('Survival Function Estimation using KM')
plt.show()
```

#### I.1.2 Cumulative Hazard Function

**Nelson-Aalen Estimator**

Non-parametric estimation of the cumulative hazard function using maximum likelihood.

\begin{equation}
    \hat{H}(t) = \sum_{i; t_i<t} \dfrac{d_i}{n_i}
\end{equation}

>with $n_i$ the number of individuals who has not experienced the event at time $t_i$

>and $d_i$ the number of individuals that experience the event at time $t_i$

```python
from lifelines import NelsonAalenFitter
naf = NelsonAalenFitter()

naf.fit(t_train, e_train)
```

```python
fig, ax = plt.subplots()
naf.plot(ax=ax)
ax.set_xlabel('t in days')
ax.set_ylabel('H_hat(t)')
ax.set_title('Cumulative Hazard Function Estimation using KM')
plt.show()
```

**Relationship between KM and NA estimate**

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))


kmf.plot(ax=ax[0])
(np.exp(-naf.cumulative_hazard_)).plot(ax=ax[0])
ax[0].set_xlabel('t in days')
ax[0].set_ylabel('S_hat(t)')
ax[0].set_title('Survival Function Estimation using KM')

(-np.log(kmf.survival_function_)).plot(ax=ax[1])
naf.plot(ax=ax[1])
ax[1].set_xlabel('t in days')
ax[1].set_ylabel('H_hat(t)')
ax[1].set_title('Cumulative Hazard Function Estimation using KM')
plt.show()
```

### Compare survival functions

#### Survival function according to gender

```python
males = (mimic['GENDER'] == 'M')

fig, ax = plt.subplots()
kmf.fit(t_train[males], e_train[males], label='male')
kmf.plot(ax=ax)
kmf.fit(t_train[~males], e_train[~males], label='female')
kmf.plot(ax=ax)

plt.title('Lifespans according to gender')
plt.show()
```

#### Log rank test

Statistical test to compare survival functions.

Hypothesis: No differences between the individuals in the probability of the event.

```python
from lifelines.statistics import logrank_test

results = logrank_test(t_train[males], t_train[~males], e_train[males], e_train[~males], alpha=.99)

results.print_summary()
```

The p_value indicates that there is no difference in the probability of death according to gender.

### Prediction using KM Estimate

Now that we have $\hat{S}(t) =P(T>t)$, we can use it to predict event for unseen patients.

$
P(T>t) = 1-P(T\leq t) \\
P(T\leq t) = 1-P(T > t)
$

Once we have the probability that the patient dies at time $t$, we can draw the event using binomial law.

```python
from sklearn.metrics import accuracy_score, f1_score
```

```python
probabilities = [kmf.predict(x) for x in t_test]
prediction = 1-np.random.binomial(1, probabilities)
```

```python
accuracy_score(e_test, prediction)
```

```python
f1_score(e_test, prediction)
```

## II. Survival Regression

Cox Regression cannot handle categorical data at the moment, so we remove them from the dataset.

```python
stamp = pd.concat([mimic[mimic.columns[0:57]], mimic['ICU']], axis=1)

stamp = stamp.dropna(axis=0)
stamp = stamp.drop(stamp.columns[stamp.dtypes == 'object'], axis=1)
```

```python
from lifelines import CoxPHFitter

stamp_train, stamp_test = train_test_split(stamp, train_size=0.8, test_size=0.2)

cph = CoxPHFitter()
cph.fit(stamp_train, duration_col='LOS', event_col='DIED', show_progress=True)

cph.print_summary()
```

### Coefficient visualization

```python
fig, ax = plt.subplots(figsize=(10,10))
cph.plot(ax=ax)
```

### Prediction Survival Function using the coefficients

Predict survival function for each individual.

```python
reg_pred = cph.predict_survival_function(stamp_test.drop(['DIED', 'LOS'], axis=1))
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
event_table = pd.Series()
for individual in stamp_test.index:
    los = stamp_test.loc[individual]['LOS']
    nearest = find_nearest(reg_pred.index.values, los)
    surv = reg_pred.loc[nearest][individual]
    serie = pd.Series(data=1-np.random.binomial(1, surv), index=pd.MultiIndex.from_tuples([individual]))
    event_table = pd.concat([event_table, serie])
```

```python
f1_score(event_table, stamp_test['DIED'])
```
